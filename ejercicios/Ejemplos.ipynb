# EJEMPLOS SPARK & SQL 

En este notebook se van a realizar los casos de uso más comunes en Databricks. Siempre que se pueda se pondrá el ejemplo en SQL y en PySpark.

## CREAR UNA TABLA

```python
%sql
CREATE OR REPLACE TABLE CLIENTES (
  cliente_id INTEGER,
  nombre STRING,
  email STRING,
  fecha_registro DATE,
  pais STRING
)
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType

# Definir esquema
schema = StructType([
    StructField("cliente_id", IntegerType(), True),
    StructField("nombre", StringType(), True),
    StructField("email", StringType(), True),
    StructField("fecha_registro", DateType(), True),
    StructField("pais", StringType(), True)
])

# Crear DataFrame vacío
df_clientes = spark.createDataFrame([], schema)

# Guardar como tabla
df_clientes.write.mode("overwrite").saveAsTable("CLIENTES")
```

## Consultar Datos de un CSV

```python
%sql
SELECT * FROM csv.`/Workspace/Users/atriguerosholgado@gmail.com/clientes.csv`
```

Como podemos observar como los csv son ficheros de tipo non-describing no es capaz de detectar que tiene cabecera el fichero. Por eso podemos utilizar la función read_files.

```python
%sql
SELECT * FROM read_files(
  '/Workspace/Users/atriguerosholgado@gmail.com/clientes.csv',
  format => 'csv',
  header => 'true',
  inferSchema => 'true'
)
```

```python
df = spark.read.csv(
    "/Workspace/Users/atriguerosholgado@gmail.com/clientes.csv",
    header=True,
    inferSchema=True           
)

df.display()
```

## Crear Tabla de un fichero CSV

```python
%sql
CREATE OR REPLACE TABLE clientes AS
SELECT * FROM read_files(
  '/Workspace/Users/atriguerosholgado@gmail.com/clientes.csv',
  format => 'csv',
  header => 'true',
  inferSchema => 'true'
)
```

```python
df = spark.read.csv(
    "/Workspace/Users/atriguerosholgado@gmail.com/clientes.csv",
    header=True,
    inferSchema=True           
)

df.write.mode("overwrite").saveAsTable("CLIENTES")

df.display()
```
